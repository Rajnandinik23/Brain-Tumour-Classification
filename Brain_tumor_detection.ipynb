```python
%%writefile app.py
import tensorflow as tf
import streamlit as st
import joblib


@st.cache(allow_output_mutation=True)
def load_model():
  model=joblib.load('/content/Brain_tumor_detction')
  return model
with st.spinner('Model is being loaded..'):
  model=load_model()

st.write("""
         # Brain Tumor Detection 
         """
         )

file = st.file_uploader("Please upload an MRI image", type=["jpg", "png"])
import cv2
from PIL import Image, ImageOps
import numpy as np
st.set_option('deprecation.showfileUploaderEncoding', False)
def import_and_predict(image_data):
        
        size = (200,200)    
        image = ImageOps.fit(image_data, size, Image.ANTIALIAS)
        image=np.expand_dims(image,axis=0)
        


        prediction = model.predict(image)
        
        return prediction
if file is None:
    st.text("Please upload an image file")
else:
    image = Image.open(file)
    st.image(image, use_column_width=True)
    op = import_and_predict(image)
    classes=['glioma_tumor','meningiomia_tumor','no_tumor','pituitary_tumor']
    score = tf.nn.softmax(op[0])
    x=classes[np.argmax(score)], 100 * np.max(score)
    st.write(x)
  

```
