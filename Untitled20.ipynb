```python
import joblib
```

```python
model=joblib.load('/content/Brain_tumor_detction')
```

```python
from PIL import Image
```

```python
import matplotlib.pyplot as plt
import numpy as np
img_path = '/content/drive/MyDrive/MRI_images/Train/glioma_tumor/gg (103).jpg'
img = plt.imread(img_path)

img=Image.open('/content/drive/MyDrive/MRI_images/Train/glioma_tumor/gg (103).jpg')
img= img.resize((200,200))
plt.imshow(img)
plt.show()
img=np.expand_dims(img,axis=0)
```

```python
op=model.predict(img)
```

```python
if (op[[0]]>=1):
      print('Glioma Tumor')
if (op[[1]]>=1):
      print('Meningioma Tumor')
if (op[[2]]>=1):
      print('No Tumor')
if (op[[3]]>=1):
      print('Pituitary Tumor')  
  
```

```python
classes=['glioma_tumor','meningiomia_tumor','no_tumor','pituitary_tumor']
import tensorflow as tf
score = tf.nn.softmax(op[0])

x=classes[np.argmax(score)], 100 * np.max(score)
x
```
